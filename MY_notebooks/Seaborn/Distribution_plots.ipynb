```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\lenovo\Desktop\Notebooks\05-Seaborn\dm_office_sales.csv')
```

```python
df
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.set(style='white')
sns.displot(data=df,x='salary',bins=10)
```

```python
#we can also use histplot function ,it is similar to displot:
sns.histplot(data=df,x='salary')
```

```python
#displot is more general so that you can make some comninations:
sns.displot(x='salary',data=df,kde=True,rug=True,color='red')
```

```python
sns.kdeplot(data=df,x='salary')
```
