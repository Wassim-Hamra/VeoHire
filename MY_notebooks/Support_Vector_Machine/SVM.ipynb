```python
from sklearn.svm import SVC
```

```python
X = df.drop('Virus Present',axis=1)
```

```python
y=df['Virus Present']
```

```python
model = SVC(kernel='linear',C=1000)
```

```python
model.fit(X,y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=0.1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=1000)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=1,gamma='scale')
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=1,gamma='auto')
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=1,gamma=0.9)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=1,gamma=2)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='sigmoid',C=1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
# as we see on the the top it is not  good
```

```python
model = SVC(kernel='poly',degree=1,C=1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly',degree=2,C=1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly',degree=2,C=1000)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = SVC()
```

```python
param_grid = {'C':[.01,.1,1],'kernel':['rbf','linear']}
```

```python
grid_model = GridSearchCV(model,param_grid=param_grid)
```

```python
grid_model.fit(X,y)
```
