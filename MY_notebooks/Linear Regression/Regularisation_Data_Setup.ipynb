```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\lenovo\Desktop\Notebooks\08-Linear-Regression-Models\Advertising.csv')
```

```python
df.head()
```

```python
# in this part w'll make the datasset to move to regularisation
```

```python
X = df.drop('sales',axis=1)
```

```python
y=df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_converter = PolynomialFeatures(degree=3,include_bias=False)
```

```python
poly_features = poly_converter.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
# Regularisation L2 : Ridge Regularisation
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train,y_train)
```

```python
test_pred = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE  = mean_absolute_error(y_test,test_pred)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_pred))
```

```python
RMSE
```

```python
#w'll use cross validation to determine the best alpha parameter : 
```

```python
#where higher is better
```

```python
from sklearn.linear_model import RidgeCV
```

```python
ridgecv_model = RidgeCV(alphas=(0.1, 1.0, 10.0))
```

```python
ridgecv_model.fit(X_train,y_train)
```

```python
ridgecv_model.alpha_
```

```python
test_pred = ridgecv_model.predict(X_test)
```

```python
MAE  = mean_absolute_error(y_test,test_pred)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test,test_pred))
```

```python
RMSE
```

```python
lassocv_model.fit(X_train,y_train)
```
