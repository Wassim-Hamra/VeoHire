```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\lenovo\Desktop\Notebooks\DATA\Advertising.csv')
```

```python
df.head()
```

```python
X  =  df.drop('sales',axis = 1)
```

```python
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_other, y_train, y_other = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_eval,X_test,y_eval,y_test = train_test_split(X_other,y_other,test_size=0.5,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_eval = scaler.transform(X_eval)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model_one = Ridge(alpha=100)
```

```python
model_one.fit(X_train,y_train)
```

```python
y_evl_pred = model_one.predict(X_eval)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_eval,y_evl_pred)
```

```python
model_two = Ridge(alpha=1)
```

```python
model_two.fit(X_train,y_train)
```

```python
y_pred = model_two.predict(X_eval)
```

```python
mean_squared_error(y_eval,y_pred)
```

```python
final_pred = model_two.predict(X_test)
```

```python
mean_squared_error(y_test,final_pred)
```
