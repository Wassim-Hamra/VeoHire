```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(1,25,2)
```

```python
y = (1/3)*x
```

```python
plt.plot(x,y);#creating the points from the two lists then math them with a line .
# when you add the ";" it is equal to "plt.show()"
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.xlim(0,20)
plt.ylim(0,7)
plt.title('Y=(1/3)*x')
plt.savefig('first_try')
```
