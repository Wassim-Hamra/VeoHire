```python
import numpy  as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,7)
b = a**4
```

```python
x=np.arange(25,40,2)
Y=2*x
```

```python
fig,axes = plt.subplots(1,2)
axes[0].plot(a,b)
axes[1].plot(x,Y)
```

```python
#if the number of subplots pass the two so we move into 2D indexing :
fig,axes = plt.subplots(2,2)
axes[0].plot(a,b)
axes[1].plot(x,Y)
```

```python
fig,axes = plt.subplots(2,2)
axes[0][0].plot(a,b)
axes[0][1].plot(x,Y)
```

```python
fig,axes = plt.subplots(3,1)
#we see that the subplots are too close so :
plt.tight_layout()
```

```python
fig,axes = plt.subplots(2,2)
axes[0][1].plot(x,Y)
axes[1][0].plot(x,Y)
axes[0][0].plot(x,Y)
axes[1][1].plot(a,b)
#we see also that they are too close :
fig.subplots_adjust(wspace=0.4,hspace=0.4)#this is the manually method we can always use "plt.tight_layout()"
axes[0][0].set_title('_____bskjgb____')#we can also make all fun learned before xlabel/xlim/...
fig.suptitle('first_subplot')#title for the hole fig
fig.savefig('first_subplot',bbox_inches='tight')
```
