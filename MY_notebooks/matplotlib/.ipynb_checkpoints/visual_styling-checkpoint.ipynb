```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lines = ax.plot(x,x,color ='blue',lw=5)
lines[0].set_dashes([1,2,1,2,10,2])
ax.plot(x,x+1,color='red',lw=5,marker='o',ms=8,markerfacecolor='green',markeredgewidth=3,markeredgecolor='yellow')
```
