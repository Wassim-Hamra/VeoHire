```python
def quality(total_bill,tip) :
    if tip/total_bill > 0.25 :
        return 'Generous'
    else :
        return 'other'
```

```python
quality(16.99,1.01)
```

```python
#we use axis=1 to meke sure we got the columnd not he rows:
df[['total_bill','tip']].apply(lambda df : quality(df['total_bill'],df['tip']),axis=1)
```

```python
#you should be careful about the structure of the function :
#so to conclude ,firstly you should grab your columns ,call apply,lambda function on df,call your custom function !!!be carefull withe the axis=1
```

```python
#what we are going to do now is making things more fast :
```

```python
#!!!!!!! _____VECTORIZE____ !!!!!! :
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```
